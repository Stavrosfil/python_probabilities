```python
import numpy as np
import matplotlib.pyplot as plot
import math
time = np.arange(0, 10, 0.1);
amplitude = np.sin(time)
# a = [i for i in range(100)]
plot.title('Sine wave')
plot.xlabel('Time')
plot.ylabel('Amplitude = sin(time)')
plot.grid(True, which='both')
# plot.axhline(y=0, color='k')
plot.plot(time, amplitude)
```

```python
from matplotlib.pyplot import figure, show
from numpy import arange, sin, pi

t = arange(0.0, 10.0, 0.001)

fig = figure(1, dpi=600)

# The numbers are: rows, collumns, and id.
ax1 = fig.add_subplot(211)
ax1.plot(t, sin(2*pi*t))
ax1.grid(True)
ax1.set_ylim((-2, 2))
ax1.set_ylabel('1 Hz')
ax1.set_title('A sine wave or two')

# for label in ax1.get_xticklabels():
#     label.set_color('r')


ax2 = fig.add_subplot(212)
ax2.plot(t, sin(2*2*pi*t))
ax2.grid(True)
ax2.set_ylim((-2, 2))
l = ax2.set_xlabel('Hi mom')
l.set_color('g')
l.set_fontsize('large')

# SAVE FILE AS IMAGE
# plot.savefig('filename.png', dpi=600)

# SAVE FILE FOR LATEX
# plot.savefig('destination_path.eps', format='eps', dpi=1000)
fig.savefig('myimage.svg', format='svg', dpi=1200)

show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
time = np.arange(0, 10, 0.1);
amplitude = np.sin(time)
# a = [i for i in range(100)]
plt.title('Beautiful sines')
plt.xlabel('Time')
plt.ylabel('Amplitude = sin(time)')
plt.grid(True, which='both')
# plot.axhline(y=0, color='k')
for i in range(50):
    plt.plot(time*i, amplitude*i*i)
    
# plt.savefig('destination_path.eps', format='eps', dpi=1000)
plt.savefig('myimage.svg', format='svg', dpi=1200)
```
