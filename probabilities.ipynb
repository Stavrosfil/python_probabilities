```python
# SIMPLE COINFLIP PROGRAM
import random
import numpy as np
import matplotlib.pyplot as plt
def coin_trial(n):
    heads = 0
    for i in range(n):
        if random.random() <= 0.5:
            heads +=1
    return heads/n*100

def simulate(numberOfTests, numberOfCoinflips):
    trials = []
    for i in range(numberOfTests):
        trials.append(coin_trial(numberOfCoinflips))
#     time = np.arange(1, n, 1);
    ans = sum(trials)/(numberOfTests)
    plt.plot(trials, 'ro')
    plt.axhline(y = ans)
    plt.xlabel('Test runs')
    plt.ylabel('Heads/run')
    plt.axis([0,numberOfTests,0,100])
    return(ans)
ansSum = 0
for i in range(1):
    ansSum += simulate(100, 100)
    
# plt.autoscale()
plt.title('Random coin flips distribution ( ' + str(ansSum) + '% )')
```

```python
#EXCERSISE 2

import random
import numpy as np
import matplotlib.pyplot as plt

# All the possible balls in the box

def chooseBalls(n = 3):
    balls = [1,2,3,4,5]
    picked = []
    for i in range(n):
        pick = random.choice(balls)
        balls.remove(pick)
        picked.append(pick)
    return picked

def simulate(n):
    trials = []
    ballsDist = []
    for i in range(n):
        picked = chooseBalls()
        ans = min(picked)
        trials.append(ans)
        for j in picked:
            ballsDist.append(j)
    f = plt.figure(1)
    plt.plot(sorted(trials), 'ro')
    plt.xlabel('Test runs')
    plt.ylabel('Result')
    plt.autoscale()
    plt.title('Random ball picks results')
    f.show()
    
    g = plt.figure(2)
    plt.plot(sorted(ballsDist))
    plt.xlabel('Test runs')
    plt.ylabel('Result')
    plt.title('Balls chosen distribution')
    g.show()
    
    print('Min is 1: ' + str(round(trials.count(1)/n*100, 2)) + '%')
    print('Min is 2: ' + str(round(trials.count(2)/n*100, 2)) + '%')
    print('Min is 3: ' + str(round(trials.count(3)/n*100, 2)) + '%')

simulate(100000)
```

```python
# EXCERSISE 3 FROM EXCERSISES SERIES 1-12

import random
import numpy as np
import matplotlib.pyplot as plt

# Initialize r
r = 10

# Initialize 2 random points on the circumference
def pickRandomPoints():
    a = random.random() * 2*r*np.pi
    b = random.random() * 2*r*np.pi
    return a, b

def run(n):
    lengths = []
    counter = 0
    for i in range(n):
        a, b = pickRandomPoints()
        L = abs(a - b)
        angle = L/r
        c = 2*r*np.sin(angle/2)
        if c < r:
            counter += 1
        lengths.append(c)
    plt.plot(sorted(lengths), 'ro')
    return counter

n = 1000000
print(str(run(n)/n*100) + '%')
```

```python
# EXCERSISE 4 FROM EXCERSISES SERIES 1-12

import random
import numpy as np
import matplotlib.pyplot as plt

def flipCoin():
    r = random.random()
    if(r > 0.5):
        return 1
    else:
        return 0
    
def runOnce():
    counter = 0
    while 1:
        ans = flipCoin()
        counter += 1
        if(ans == 1):
            return (counter - 1) % 3

def simulate(n):
    wins = [0] * 3
    for i in range(n):
        wins[runOnce()] += 1
    plt.plot(wins, 'ro')
    print(wins)
    
simulate(1000000)
```

```python
# EXCERSISE 6 FROM EXCERSISES SERIES 1-12
# The bitrhday problem
import random
import numpy as np
import matplotlib.pyplot as plt

def calculate(n):
    notTheSameDay = 1
    for i in range(n):
        notTheSameDay *= (365 - i)/365
    # print(notTheSameDay)
    return notTheSameDay
    
def plotProbs(n):
    sameDayProb = []
    for i in range(n):
        sameDayProb.append(1 - calculate(i))
    plt.figure(dpi=100)
    plt.plot(sameDayProb, 'ro')
    
plotProbs(60)
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

def xiStinx(n):
    ans = []
    for i in range(n):
        ans.append(np.power(i, i))
    return ans

def paragontiko(n):
    ans = []
    for i in range(n):
        ans.append(np.math.factorial(i))
    return ans

plt.semilogy(paragontiko(10))
plt.semilogy(xiStinx(10))
```
