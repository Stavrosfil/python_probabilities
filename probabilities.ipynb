```python
# SIMPLE COINFLIP PROGRAM
import random
import numpy as np
import matplotlib.pyplot as plt
def coin_trial(n):
    heads = 0
    for i in range(n):
        if random.random() <= 0.5:
            heads +=1
    return heads

def simulate(numberOfTests, numberOfCoinflips):
    trials = []
    for i in range(numberOfTests):
        trials.append(coin_trial(numberOfCoinflips))
#     time = np.arange(1, n, 1);
    ans = sum(trials)/(numberOfTests)
    plt.plot(trials, 'ro')
    plt.axhline(y = ans)
    plt.xlabel('Test runs')
    plt.ylabel('Heads/run')
    plt.axis([0,numberOfTests,0,numberOfCoinflips])
    return(ans)
ansSum = 0
for i in range(1):
    ansSum += simulate(100, 100)
    
plt.autoscale()
plt.title('Random coin flips distribution ( ' + str(ansSum) + '% )')
```

```python
#EXCERSISE 2

import random
import numpy as np
import matplotlib.pyplot as plt

# All the possible balls in the box

def chooseBalls(n = 3):
    balls = [1,2,3,4,5]
    picked = []
    for i in range(n):
        pick = random.choice(balls)
        balls.remove(pick)
        picked.append(pick)
    return picked

def simulate(n):
    trials = []
    ballsDist = []
    for i in range(n):
        picked = chooseBalls()
        ans = min(picked)
        trials.append(ans)
        for j in picked:
            ballsDist.append(j)
    f = plt.figure(1)
    plt.plot(sorted(trials), 'ro')
    plt.xlabel('Test runs')
    plt.ylabel('Result')
    plt.autoscale()
    plt.title('Random ball picks results')
    f.show()
    
    g = plt.figure(2)
    plt.plot(sorted(ballsDist))
    plt.xlabel('Test runs')
    plt.ylabel('Result')
    plt.title('Balls chosen distribution')
    g.show()
    
    print('Min is 1: ' + str(round(trials.count(1)/n*100, 2)) + '%')
    print('Min is 2: ' + str(round(trials.count(2)/n*100, 2)) + '%')
    print('Min is 3: ' + str(round(trials.count(3)/n*100, 2)) + '%')

simulate(10000)
```

```python
# EXCERSISE 6 FROM EXCERSISES SERIES 1-12
# The bitrhday problem
import random
import numpy as np
import matplotlib.pyplot as plt

def calculate(n):
    notTheSameDay = 1
    for i in range(n):
        notTheSameDay *= (365 - i)/365
    # print(notTheSameDay)
    return notTheSameDay
    
def plotProbs(n):
    sameDayProb = []
    for i in range(n):
        sameDayProb.append(1 - calculate(i))
    plt.figure(dpi=150)
    plt.plot(sameDayProb, 'ro')
    
plotProbs(60)
```
